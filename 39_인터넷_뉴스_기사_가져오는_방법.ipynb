```python
import pandas as pd
from bs4 import BeautifulSoup
import time
import requests
```

```python
total = []

url = requests.get("https://search.daum.net/search?w=news&nil_search=btn&DA=PGD&enc=utf8&cluster=y&cluster_page=1&q=%ED%95%9C%EA%B5%AD&p=1")
html = BeautifulSoup(url.text)



news = html.find("ul", class_='c-list-basic').find_all('div', class_='c-item-content')
for i in news:
    title = i.find('div', class_ = 'item-title').text
    content = i.find('p', class_='conts-desc clamp-g2').text

    total.append([title, content])
len(total)
```

```python
df = pd.DataFrame(total, columns=['제목', '요약'])
df
```
