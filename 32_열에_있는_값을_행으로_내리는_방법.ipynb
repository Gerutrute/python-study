```python
import pandas as pd

df=pd.read_csv("pew.csv")
df2=pd.melt(df, id_vars=['religion'], var_name='income', value_name='count')
df2
```

```python
billboard=pd.read_csv("billboard.csv")
billboard2=pd.melt(billboard, id_vars=['year', 'artist', 'track', 'time', 'date.entered'],
                   var_name='week', value_name='rating')
billboard2
```

```python
ebola=pd.read_csv("country_timeseries.csv")
ebola
```

```python
ebola2=pd.melt(ebola, id_vars=['Date','Day'])#var, value name이 지정되지 않은 경우, variable과 value로 알아서 지정됨.
ebola2
```

```python
ebola_split=ebola2['variable'].str.split("_")#_를 기줌으로 자름
ebola_split
```

```python
state=ebola_split.str[0]
country=ebola_split.str[1]

ebola2['State']=state
ebola2['Country']=country

del ebola2['variable']
ebola2
```
