```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

tips = sns.load_dataset("tips")
tips
```

```python
fig = plt.figure()#그림을 그리기 위한 도화지를 fig변수에 할당

ax1 = fig.add_subplot(1,1,1) #1행 1열 1번쨰 자리에 할당
ax1.hist(tips['total_bill'], bins=10)
```

```python
ax1.hist(tips['total_bill'], bins=30)

fig
```

```python
fig = plt.figure()#그림을 그리기 위한 도화지를 fig변수에 할당
ax1 = fig.add_subplot(1,1,1) #1행 1열 1번쨰 자리에 할당

ax1.scatter(tips['total_bill'], tips['tip'], c=tips['size']#사이즈별로 색을 다르게함
            , s=tips['size']*20#개체 수치만큼 크게함
            , alpha=0.5#투명도 조정
            )
ax1.set_xlabel("TOTAL BILL")
ax1.set_ylabel("TIP")
ax1.set_title("SCATTER")
```

```python
fig = plt.figure()#그림을 그리기 위한 도화지를 fig변수에 할당
ax1 = fig.add_subplot(1,1,1) #1행 1열 1번쨰 자리에 할당

male = tips[tips['sex']=='Male']
female = tips[tips['sex']=='Female']
ax1.set_title("GENDER TIP")#표 제목
ax1.set_xlabel("Gender")#x축 제목
ax1.set_ylabel("Bill")#y축 제목
ax1.boxplot([male['total_bill'], female['total_bill']])
```
